In this task, we will implement an Apache Kafka consumer to consume the data from task
2.7a.

```python
# import statements
from time import sleep
from kafka import KafkaConsumer
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import stream data from kafka
topic = 'prediction_of_count_new'


def connect_kafka_consumer():
    _consumer = None
    try:
        _consumer = KafkaConsumer( topic,
                                  consumer_timeout_ms=10000,  # stop iteration if no message after 10 sec
                                  bootstrap_servers=['localhost:9092'],
                                  api_version=(0, 10))
    except Exception as ex:
        print('Exception while connecting Kafka')
        print(str(ex))
    finally:
        return _consumer
```

```python
# convert the data
consumer = connect_kafka_consumer()

col_list = ['senser_name','latitude','longitude']
data_list = []

for message in consumer:
    data = str(message.value.decode('utf-8')).strip('[ ]').split(',')
    data_list.append(data)
df = pd.DataFrame(data_list, columns = col_list)
```
